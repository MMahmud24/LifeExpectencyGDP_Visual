# Life Expectancy V.S. GDP in Several Different Countries

In this project, we will be looking at relationships between GDP(Gross Domestic Product) and life expectancy in 6 different countries from 2000 to 2015 using data visualizations and statistical reasoning.

The 6 countries are:
- Chile
- China
- Germany
- Mexico
- The United States
- Zimbabwe

First, lets take a statistical analysis of GDP and life expectancy of these different countries by using measures of central tendency

```python
import pandas as pd
import numpy as np
import matplotlib as plt

# Load the data
data = pd.read_csv('all_data.csv')

# Mean GDP(In Billions) of Each Country
data = data.assign(GDP_in_Billions=data['GDP'] / 1_000_000_000)
mean_gdp = data.groupby('Country').GDP_in_Billions.mean()
print(mean_gdp)

#Mean Life Expectancy of Each Country 
mean_life_expectancy = data.groupby('Country').Life_Expectancy.mean()
print(mean_life_expectancy)
```

From this data, we can infer that GDP and life expectancy could be slightly associated with one another. Let's look at Zimbabwe, the mean GDP is much lower than the rest of the countries and so is the life expectancy. However, this isn't enough to prove the association since there are some present inconsistencies. Let's look at Chile and China. Chile has a much lower GDP but a greater life expectancy. This same pattern is seen with Germany and the United States. So from this data alone, we cannot conclude a possible association between GDP and life expectancy.

However, let's look at each country individually.

## Year V.S. GDP

```python
chile_data = data[data.Country == 'Chile']
china_data = data[data.Country == 'China']
germany_data = data[data.Country == 'Germany']
mexico_data = data[data.Country == 'Mexico']
us_data = data[data.Country == 'United States of America']
zimbabwe_data = data[data.Country == 'Zimbabwe']

#plotting GDP for Each Country Over the Years
plt.figure(figsize=(12, 8)) 
plt.suptitle('GDP For Each Country Over the Years', fontsize=16)

chile_graph1 = plt.subplot(3,2,1)
chile_graph1.scatter(chile_data.Year, chile_data.GDP_in_Billions)
chile_graph1.set_title('Chile')
chile_graph1.set_xlabel('Year')
chile_graph1.set_ylabel('GDP in Billions')

china_graph = plt.subplot(3,2,2)
china_graph.scatter(china_data.Year, china_data.GDP_in_Billions)
china_graph.set_title('China')
china_graph.set_xlabel('Year')
china_graph.set_ylabel('GDP in Billions')

germany_graph = plt.subplot(3,2,3)
germany_graph.scatter(germany_data.Year, germany_data.GDP_in_Billions)
germany_graph.set_title('Germany')
germany_graph.set_xlabel('Year')
germany_graph.set_ylabel('GDP in Billions')

mexico_graph = plt.subplot(3,2,4)
mexico_graph.scatter(mexico_data.Year, mexico_data.GDP_in_Billions)
mexico_graph.set_title('Mexico')    
mexico_graph.set_xlabel('Year')
mexico_graph.set_ylabel('GDP in Billions')

us_graph = plt.subplot(3,2,5)
us_graph.scatter(us_data.Year, us_data.GDP_in_Billions)
us_graph.set_title('United States of America')
us_graph.set_xlabel('Year')
us_graph.set_ylabel('GDP in Billions')

zimbabwe_graph = plt.subplot(3,2,6)
zimbabwe_graph.scatter(zimbabwe_data.Year, zimbabwe_data.GDP_in_Billions)
zimbabwe_graph.set_title('Zimbabwe')
zimbabwe_graph.set_xlabel('Year')
zimbabwe_graph.set_ylabel('GDP in Billions')

plt.tight_layout()
plt.show()
```

We can see from these graphs that GDP has gone up from 2000 to 2015 for every country. From this we can conclude that as time progressed, so did a country's GDP. Time and GDP have a positive correlation. 

Let's take a look at life expectancy over the years. 

## Year V.S. Life Expectancy

```python
plt.figure(figsize=(12, 8))
plt.suptitle('Life Expectancy For Each Country Over the Years', fontsize=16)

chile_graph2 = plt.subplot(3,2,1)
chile_graph2.scatter(chile_data.Year, chile_data.Life_Expectancy)
chile_graph2.set_title('Chile')
chile_graph2.set_xlabel('Year')
chile_graph2.set_ylabel('Life Expectancy')

china_graph2 = plt.subplot(3,2,2)
china_graph2.scatter(china_data.Year, china_data.Life_Expectancy)
china_graph2.set_title('China')
china_graph2.set_xlabel('Year')
china_graph2.set_ylabel('Life Expectancy')

germany_graph2 = plt.subplot(3,2,3)
germany_graph2.scatter(germany_data.Year, germany_data.Life_Expectancy)
germany_graph2.set_title('Germany')
germany_graph2.set_xlabel('Year')
germany_graph2.set_ylabel('Life Expectancy')

mexico_graph2 = plt.subplot(3,2,4)
mexico_graph2.scatter(mexico_data.Year, mexico_data.Life_Expectancy)
mexico_graph2.set_title('Mexico')
mexico_graph2.set_xlabel('Year')
mexico_graph2.set_ylabel('Life Expectancy')

us_graph2 = plt.subplot(3,2,5)
us_graph2.scatter(us_data.Year, us_data.Life_Expectancy)
us_graph2.set_title('United States of America')
us_graph2.set_xlabel('Year')
us_graph2.set_ylabel('Life Expectancy')

zimbabwe_graph2 = plt.subplot(3,2,6)
zimbabwe_graph2.scatter(zimbabwe_data.Year, zimbabwe_data.Life_Expectancy)
zimbabwe_graph2.set_title('Zimbabwe')
zimbabwe_graph2.set_xlabel('Year')
zimbabwe_graph2.set_ylabel('Life Expectancy')

plt.tight_layout()
plt.show()
```

Again, we see that for all of the countries, their average life expectancies all have gone from 2000 to 2015. We can conclude from this data that time and life expectancy also have a positive correlation.

Now, lets look at the relationship between GDP and Life Expectency for each country. 

## GDP V.S. Life Expectancy

```python
#graphing GDP vs Life_Expectancy for each country using scatterscatters

plt.figure(figsize=(12, 8))
plt.suptitle('Relationship between GDP and Life Expectancy For Each Country', fontsize=16)

chile_graph3 = plt.subplot(3,2,1)
chile_graph3.scatter(chile_data.GDP_in_Billions, chile_data.Life_Expectancy)
chile_graph3.set_title('Chile')
chile_graph3.set_xlabel('GDP in Billions')
chile_graph3.set_ylabel('Life Expectancy')

china_graph3 = plt.subplot(3,2,2)
china_graph3.scatter(china_data.GDP_in_Billions, china_data.Life_Expectancy)
china_graph3.set_title('China')
china_graph3.set_xlabel('GDP in Billions')
china_graph3.set_ylabel('Life Expectancy')

germany_graph3 = plt.subplot(3,2,3)
germany_graph3.scatter(germany_data.GDP_in_Billions, germany_data.Life_Expectancy)
germany_graph3.set_title('Germany')
germany_graph3.set_xlabel('GDP in Billions')
germany_graph3.set_ylabel('Life Expectancy')

mexico_graph3 = plt.subplot(3,2,4)
mexico_graph3.scatter(mexico_data.GDP_in_Billions, mexico_data.Life_Expectancy)
mexico_graph3.set_title('Mexico')
mexico_graph3.set_xlabel('GDP in Billions')
mexico_graph3.set_ylabel('Life Expectancy')

us_graph3 = plt.subplot(3,2,5)
us_graph3.scatter(us_data.GDP_in_Billions, us_data.Life_Expectancy)
us_graph3.set_title('United States of America')
us_graph3.set_xlabel('GDP in Billions')
us_graph3.set_ylabel('Life Expectancy')

zimbabwe_graph3 = plt.subplot(3,2,6)
zimbabwe_graph3.scatter(zimbabwe_data.GDP_in_Billions, zimbabwe_data.Life_Expectancy)
zimbabwe_graph3.set_title('Zimbabwe')
zimbabwe_graph3.set_xlabel('GDP in Billions')
zimbabwe_graph3.set_ylabel('Life Expectancy')

plt.tight_layout()
plt.show()
```

From these graphs, we can conclude a positive correlation between GDP and life expectancy for each country. Therefore, we can infer that these two variables are associated. 

# Conclusion

In this analysis, we have looked at the possible association between GDP and life expectancy for 6 different countries. At first, by looking at all 5 countries as a whole, we evaluated that GDP isn't necessarily associated with life expectancy. This was seen by comparing the mean GDP and life expectancies throughout 15 years for all the countries. A higher GDP did not necessarily mean a higher life expectancy from this data alone. 

I decided to take a closer look at each country individually through data visualization. My choice of scatterplots was a result of the variables being quantitative variables, where using a scatterplot would lead to the best visualization of these variables.

First, we took a look at Year V.S. GDP. Typically, as life expectancy rose, so did GDP. This proves a positive correlation between time and GDP. Next, we looked at Year V.S. Life expectancy where, similarly, we also noticed a positive correlation between time and life expectancy. Lastly, we looked at GDP V.S. life expectency where again we saw another positive correlation.

Just from observing the graphs, we can infer that there is a possible association between GDP and life expectancy. If we look at the Zimbabwe's graphs, we notice that the dip in GDP and life expectency occur around the same time frame, which is the early 2000s. While this is a sole case, it cannot be ignored. 

All in all, it can be concluded from these tests that there may be an association between GDP and life expectency. A country's GDP can evaluate a country's ability to provide the essential needs and services to its citizens. A higher GDP could signify a higher availability of health care within the country, which could lead to higher life expectancies since the civilians would have access to health care and medical services when feeling ill.
